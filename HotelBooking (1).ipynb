```python
def find_outliers_iqr(df):
    outliers_dict = {}
    
    for column in df_numerical.columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        outliers_count = outliers.shape[0]
        outliers_dict[column] = outliers_count
    
    return outliers_dict

outliers_dict = find_outliers_iqr(df)
outliers_dict
```

```python
def find_outliers_per(df):
    outliers_dict1 = {}
    
    for column in df_numerical.columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        outliers_per = len(outliers)/len(df) * 100
        outliers_dict1[column] = outliers_per
    
    return outliers_dict1

outliers_dict1 = find_outliers_per(df)
outliers_dict1
```

```python
def plotvariable(df, variable):
    plt.figure(figsize=(10,3))
    plt.subplot(1,2,1)
    plt.hist(df[variable])
    plt.title(f"Histogram of {variable}")
    plt.subplot(1,2,2)
    sns.boxplot(df[variable])
    plt.title(f"Boxplot of {variable}")
    plt.show()

for i in df_numerical.columns:
    plotvariable(df,i)
    
    
```

```python
def handle_outliers(df, columns,method):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        if method == 'remove':
            df = df[~((df[col] < lower_bound) | (df[col] > upper_bound))]
        elif method == 'replace':
            median = df[col].median()
            df[col] = np.where((df[col] < lower_bound) | (df[col] > upper_bound), median, df[col])
    
    return df

# Apply outlier handling
columns_to_replace = ['lead_time', 'adr', 'total_of_special_requests','previous_bookings_not_canceled','required_car_parking_spaces', 'children','booking_changes','adults']
columns_to_remove = ['stays_in_weekend_nights', 'stays_in_week_nights','days_in_waiting_list', 'babies', 'previous_cancellations']

df_cleaned = handle_outliers(df.copy(), columns_to_remove,method = 'remove')
df_cleaned = handle_outliers(df_cleaned, columns_to_replace, method='replace')

print("Cleaned DataFrame:")
df_cleaned
```

# Insights and Suggestions
1. From the above analysis we can observe that 27.5 % bookings are canceled. To reduce the percentage of cancellations , first we have to find the reasons behind them.
2. We observed that If the average number of days in the waiting list is higher, there is a greater chance that bookings will be canceled. So Reducing waiting times will decrease the number of cancellations.
3. Next we observed a higher count of transient bookings which means most people are staying for short periods. So Hotels have to create special deals for short stays, like discounts for 1-3 night bookings, and benefits like free breakfast ,free WI-FI or a city tour. This can attract more Customers who prefer staying for shorter periods.
4. If the hotels offer special summer discounts or packages during May, June then they can attract more customers. And during off-peak seasons like november, december, january the hotels should host events, festivals, or other activities to attract more visitors 
5. Hotels should provide rewards and discounts to the repeated customers.This can encourage customers to choose the hotel again.
6. We observed that online travel agencies are the main source of bookings. So hotels should invest in improving their online travel agency for best user experiences and should provide special offers for more bookings. 
7. Hotels should implement a referral program where existing customers can refer friends and family in exchange for discounts or rewards.
Promote the referral program through email marketing and social media channels.
8. By implementing the above suggestions, hotels can reduce booking cancellations and increase the number of reservations.
